# 1. Predict Billboard Chart 2019

### 주간 차트를 활용하여 연간 차트 추측하기

billboard에서 제공하는 주간 차트를 통해 연간 차트를 예측해볼 수 있는 코드입니다. 주석을 참고하면서 작동시키길 바랍니다.

```python
# Operating code 1
# 2019년 첫 주 부터 마지막 주 까지의 날짜에 대한 데이터 얻기

import datetime

start_of_2019=datetime.date(2019,1,1)
nextweek=start_of_2019
daylist_of_2019=[]
i=-1

while nextweek.year==2019:
    i+=1
    nextweek=start_of_2019+datetime.timedelta(weeks=i)
    daylist_of_2019.append(str(nextweek))
    
daylist_of_2019.pop()
print(daylist_of_2019)
```

```python
# Operating code 2
# billboard.com에서 2019년 데이터 크롤링 하기
# 간혹 크롤링 하는 도중 처리과정이 갑자기 빠르게 진행되며 크롤링이 정상적으로 되지 않을 때가 있습니다. 그럴 때는 아래의 코드를 참고하길 바랍니다.
# 일반적으로 저장된 csv 파일의 크기는 7~8KB이며, 1KB는 크롤링이 되지 않은 것에 해당합니다.
 
from bs4 import BeautifulSoup
import requests
import re
import csv
import os

os.makedirs('Let`s_predict_billboard_chart')

for i in range(len(daylist_of_2019)):
    
    csv_filename = ".//Let`s_predict_billboard_chart//billboard_chart_2019년_%d번째_주_순위.csv"%(i+1)
    crawling_url = "https://www.billboard.com/charts/hot-100/%s"% daylist_of_2019[i]
    csv_open = open(csv_filename,"w+",encoding = 'utf-8')
    csv_writer = csv.writer(csv_open)
    csv_writer.writerow(('rank','song','artist','song_information','point'))
    response = requests.get(crawling_url)
    bs = BeautifulSoup(response.text,'html.parser')
    chart_list = bs.find_all('li',{"class":re.compile('chart-list__elements*')})
    print("곡에 대한 정보 추출 %d%% 진행 중"%(i/len(daylist_of_2019)*100))
    
    for chart in chart_list:
        rank  = chart.find_all('span',{"class":re.compile('chart-element__rank__number')})
        real_rank = int(rank[0].text)
        song  = chart.find_all('span',{"class":re.compile('chart-element__information__song text--truncate color--primary')})
        real_song = song[0].text
        artist = list(chart.find_all('span',{"class":re.compile('chart-element__information__artist text--truncate color--secondary')}))
        real_artist = artist[0].text
        song_information=real_song+'/'+real_artist
        point = str(101 - real_rank)
        real_point=point
        csv_writer.writerow((real_rank,real_song,real_artist,song_information,real_point))
    csv_open.close()
    
print("곡에 대한 정보 추출이 완료됨")
```

```python
# operating code 2.1
# 위의 코드에서 오류가 발생했다면 이 코드를 실행시키길 바랍니다.
# 이 코드를 또 오류가 나면 다시 오류가 난 주차를 입력하십시오.
# 오류가 계속 발생하면 모든 커널을 초기화 시키고 작동하시길 바랍니다.

from bs4 import BeautifulSoup
import requests
import re
import csv
import os

#error=int(input("오류가 발생하지 않았다면 숫자 0을 입력하시오. 몇 주차부터 오류가 발생했나요? 1~53사이의 숫자를 입력하길 바랍니다."))

while True:
    error=input("오류가 발생하지 않았다면 이 다음 코드를 실행하세요. 몇 주차부터 오류가 발생했나요? 1~53사이의 숫자를 입력하길 바랍니다.")
    if error.isdigit()==False:
        continue
    elif error.isdigit()==True:
        if int(error) not in range(1,54):
            continue
        elif int(error) in range(1,54):
            break

for i in range(int(error)-1,len(daylist_of_2019)):
    
    csv_filename = ".//predict_billboard_chart//billboard_chart_2019년_%d번째_주_순위.csv"%(i+1)
    crawling_url = "https://www.billboard.com/charts/hot-100/%s"% daylist_of_2019[i]
    csv_open = open(csv_filename,"w+",encoding = 'utf-8')
    csv_writer = csv.writer(csv_open)
    csv_writer.writerow(('rank','song','artist','song_information','point'))
    response = requests.get(crawling_url)
    bs = BeautifulSoup(response.text,'html.parser')
    chart_list = bs.find_all('li',{"class":re.compile('chart-list__elements*')})
    print("곡에 대한 정보 추출 %d%% 진행 중"%(i/len(daylist_of_2019)*100))
    
    for chart in chart_list:
        rank  = chart.find_all('span',{"class":re.compile('chart-element__rank__number')})
        real_rank = int(rank[0].text)
        song  = chart.find_all('span',{"class":re.compile('chart-element__information__song text--truncate color--primary')})
        real_song = song[0].text
        artist = list(chart.find_all('span',{"class":re.compile('chart-element__information__artist text--truncate color--secondary')}))
        real_artist = artist[0].text
        song_information=real_song+'/'+real_artist
        point = str(101 - real_rank)
        real_point=point
        csv_writer.writerow((real_rank,real_song,real_artist,song_information,real_point))
    csv_open.close()
    
print("곡에 대한 정보 추출이 완료됨")
```

```python
# Operating code 3
# 각 노래마다 얻은 포인트 측정하기

import pandas as pd
import numpy as np

year_chart_list=[]
year_chart_dict={}

for i in range(1,len(daylist_of_2019)):
    df1=pd.read_csv(".//Let`s_predict_billboard_chart//billboard_chart_2019년_%d번째_주_순위.csv"% i,engine='python')
    songinfo=df1.loc[:,['song_information']]
    songpoint=df1.loc[:,['point']]
    
    songinfolist=np.array(df1['song_information'].tolist())
    pointlist=np.array(df1['point'].tolist())

    info_point_dict=dict(zip(songinfolist,pointlist))
    
    infolist=list(info_point_dict.keys())
    pointlist=list(info_point_dict.values())
    
    for j in range(len(infolist)):
        if infolist[j] not in year_chart_list:
            year_chart_list.append(infolist[j])
            year_chart_dict[infolist[j]]=pointlist[j]         
        elif infolist[j] in year_chart_list:
            pointlist[j]+=year_chart_dict.get(infolist[j])
            year_chart_dict[infolist[j]]=pointlist[j]


orderly2=sorted(year_chart_dict.items(), key=lambda t : t[1], reverse=True)

print(orderly2)
print("2019년 billboard chart에 올라간 곡의 갯수: "+str(len(orderly2)))
```

```python
# operating code 4
# 주간 차트를 활용하여 예측한 상위 100곡 직접 확인하기

predicted_top_100_list=[]

for i in range(0,100):
    predicted_top_100_list.append(orderly2[i])
    
for i in range(0,100):
    print(i+1,orderly2[i])
```

```python
# operating code 5
# billboard에서 제공하는 2019년 차트를 csv파일로 저장하기

from bs4 import BeautifulSoup
import requests
import re
import csv
import os

csv_filename = "official_billboard_chart_2019_year_end.csv"
crawling_url = "https://www.billboard.com/charts/year-end/2019/hot-100-songs"
csv_open = open(csv_filename, "wt", encoding = 'utf-8')
csv_writer = csv.writer(csv_open)
csv_writer.writerow(('rank of 2019','song_information'))
response = requests.get(crawling_url)
bs = BeautifulSoup(response.text, 'html.parser')
chart_list = bs.find_all('div', {"class":'ye-chart-item__primary-row'})

for chart in chart_list:
    rank = chart.find('div',{"class":'ye-chart-item__rank'})
    real_rank = rank.text
    song = chart.find('div',{"class":'ye-chart-item__title'})
    real_song = song.text
    artist =chart.find('div',{"class":'ye-chart-item__artist'})
    real_artist = artist.text
    song_information=real_song+'/'+real_artist
    csv_writer.writerow((real_rank,song_information))
    
csv_open.close()
```

```python
# operating code 6
# 크롤링을 통해 예측한 상위 100곡에 대한 데이터를 csv파일로 저장하기

import csv

f=open('predicted_2019_billboard_chart_year_end.csv','w',newline='')
wr=csv.writer(f)
wr.writerow(("Predicted Rank","Predicted Song Information"))
wr.writerow(("",""))

for i in range(0,len(predicted_top_100_list)):
    wr.writerow(("%d"%(i+1) ,str(predicted_top_100_list[i][0])))
    wr.writerow(("",""))
    

f.close()
```
